### Анализ распределений признаков и исследование зависимости признаков от Churn


Функция для автоматического анализа признака и исследование зависимости признаков от Churn

Эта функция и некоторые другие функции по меньше я использовал и в прошлых работ они являются универсальными

```python
def plot_feature_analysis(df, feature, target='default', alpha=0.05):
    """
    Автоматический анализ признака:
      • Определяет тип (binary, categorical, numerical) по данным
      • Показывает количество NaN
      • Строит 3 графика: распределение, зависимость от target, boxplot/heatmap
      • Для pie-chart добавляет под процентом реальное количество
      • Для bar-chart зависимости от target добавляет точное число в категории
      • Выполняет статистический тест и сразу пишет, есть ли статистически значимая связь:
         – Для числовых: Welch’s t-test между группами Churn=0 и Churn=1
         – Для категориальных/бинарных: χ²-тест независимости
      :param df: DataFrame с данными (должен содержать feature и target)
      :param feature: Строка, название анализируемого столбца
      :param target: Строка, название целевой переменной (по умолчанию 'Churn')
      :param alpha: Уровень значимости (по умолчанию 0.05)
    """
    # Проверка наличия колонок
    if feature not in df.columns or target not in df.columns:
        raise ValueError(f"В DataFrame нет столбца '{feature}' или '{target}'")
    
    # Подсчёт пропусков
    nan_count = df[feature].isna().sum()
    total = len(df)
    print(f"\n=== Анализ '{feature}' (пропусков: {nan_count}/{total}, {nan_count/total:.1%}) ===")
    
    # Автоопределение типа признака
    unique_vals = df[feature].dropna().unique()
    nunique = len(unique_vals)
    if nunique == 2:
        feature_type = 'binary'
    elif is_numeric_dtype(df[feature]) and nunique > 10:
        feature_type = 'numerical'
    else:
        feature_type = 'categorical'
    
    # Вспомогательная функция для pie: показывает и % и точное количество
    def make_autopct(values):
        def my_autopct(pct):
            total_vals = np.sum(values)
            count = int(round(pct * total_vals / 100.0))
            return f"{pct:.1f}%\n({count})"
        return my_autopct
    
    # Подготовка фигуры
    plt.figure(figsize=(14, 10))
    gs = GridSpec(2, 2, height_ratios=[1, 1])
    plt.suptitle(f"Анализ признака: {feature}", fontsize=20, fontweight='bold', y=0.97)
    
    # Распределение признака (верхний левый)
    ax1 = plt.subplot(gs[0, 0])
    if feature_type == 'binary':
        counts = df[feature].value_counts(dropna=False)
        wedges, texts, autotexts = ax1.pie(
            counts,
            labels=counts.index.astype(str),
            autopct=make_autopct(counts.values),
            colors=['#4e79a7', '#f28e2b'],
            startangle=90,
            textprops={'fontsize': 12, 'color': 'white'},
            wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
        )
        ax1.set_title(f"Распределение {feature}", fontsize=14)
        ax1.legend(
            wedges,
            counts.index.astype(str),
            title=feature,
            loc='center left',
            bbox_to_anchor=(1, 0.5),
            fontsize=12
        )
    
    elif feature_type == 'categorical':
        sns.countplot(x=feature, data=df, ax=ax1, palette='viridis')
        ax1.set_title(f"Распределение {feature}", fontsize=14)
        ax1.set_xlabel("")
        total = len(df)
        for p in ax1.patches:
            height = p.get_height()
            ax1.annotate(
                f"{height/total:.1%}\n({height})",
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center',
                xytext=(0, 7), textcoords='offset points',
                fontsize=11
            )
    
    else:  # numerical
        sns.histplot(df[feature].dropna(), kde=True, ax=ax1, color='#2c3e50', bins=20)
        mean_val = df[feature].mean()
        ax1.axvline(mean_val, color='#e74c3c', linestyle='--',
                    label=f"Среднее: {mean_val:.2f}")
        ax1.set_title(f"Распределение {feature}", fontsize=14)
        ax1.set_xlabel(feature, fontsize=12)
        ax1.set_ylabel("Плотность", fontsize=12)
        ax1.legend()
    
    # Зависимость от target (верхний правый) + статистический тест
    ax2 = plt.subplot(gs[0, 1])
    if feature == target:
        ax2.text(0.5, 0.5, "Нельзя строить зависимость от самого себя",
                 ha='center', va='center', fontsize=14, color='gray')
        ax2.axis('off')
    
    elif feature_type in ['binary', 'categorical']:
        raw_ctab = pd.crosstab(df[feature], df[target])
        ctab = pd.crosstab(df[feature], df[target], normalize='index').fillna(0)
        
        cat_values = ctab.index.tolist()
        cat_labels = [str(v) for v in cat_values]
        x = np.arange(len(cat_values))
        bar_w = 0.35
        
        loyal = ctab.get(0, pd.Series(0, index=ctab.index))
        churn = ctab.get(1, pd.Series(0, index=ctab.index))
        
        bars0 = ax2.bar(x, loyal, bar_w, color='#4caf50', label='Лояльные')
        bars1 = ax2.bar(x + bar_w, churn, bar_w, color='#e74c3c', label='Отток')
        ax2.set_title(f"Доля оттока по категориям {feature}", fontsize=14)
        ax2.set_xticks(x + bar_w / 2)
        ax2.set_xticklabels(cat_labels, rotation=45, ha='right')
        ax2.set_ylabel("Доля", fontsize=12)
        ax2.legend()
        
        # Подписи: процент + точное число
        for i, val in enumerate(cat_values):
            count_loyal = raw_ctab.loc[val].get(0, 0)
            count_churn = raw_ctab.loc[val].get(1, 0)
            pct_loyal = loyal.iloc[i]
            pct_churn = churn.iloc[i]
            ax2.annotate(
                f"{pct_loyal:.1%}\n({count_loyal})",
                (bars0[i].get_x() + bars0[i].get_width() / 2, pct_loyal),
                ha='center', va='bottom',
                xytext=(0, 3), textcoords='offset points',
                fontsize=10
            )
            ax2.annotate(
                f"{pct_churn:.1%}\n({count_churn})",
                (bars1[i].get_x() + bars1[i].get_width() / 2, pct_churn),
                ha='center', va='bottom',
                xytext=(0, 3), textcoords='offset points',
                fontsize=10
            )
        
        # χ²-тест независимости
        try:
            chi2_stat, chi2_p, _, _ = chi2_contingency(raw_ctab, correction=False)
            if chi2_p < alpha:
                print(f"χ²-тест: χ² = {chi2_stat:.3f}, p-value = {chi2_p:.3f} → связь есть (p < {alpha})")
            else:
                print(f"χ²-тест: χ² = {chi2_stat:.3f}, p-value = {chi2_p:.3f} → нет статистически значимой связи (p ≥ {alpha})")
        except Exception:
            print(f"χ²-тест не выполнен (недостаточно данных).")
    
    else:  # numerical
        sns.kdeplot(
            data=df, x=feature, hue=target, ax=ax2,
            palette=['#4caf50', '#e74c3c'],
            fill=True, alpha=0.5, common_norm=False
        )
        ax2.set_title(f"{feature} в разрезе {target}", fontsize=14)
        ax2.set_xlabel(feature, fontsize=12)
        ax2.set_ylabel("Плотность", fontsize=12)
        ax2.legend(title=target, labels=['Лояльные', 'Ушедшие'])
        
        grp0 = df.loc[df[target] == 0, feature].dropna()
        grp1 = df.loc[df[target] == 1, feature].dropna()
        if len(grp0) >= 3 and len(grp1) >= 3:
            t_stat, p_val = ttest_ind(grp0, grp1, equal_var=False, nan_policy='omit')
            if p_val < alpha:
                print(f"Welch’s t-test: t = {t_stat:.3f}, p = {p_val:.3f} → средние различаются (p < {alpha})")
            else:
                print(f"Welch’s t-test: t = {t_stat:.3f}, p = {p_val:.3f} → нет значимой разницы (p ≥ {alpha})")
        else:
            print("t-test не выполнен (в одной из групп меньше 3 наблюдений).")
    
    # Boxplot или Heatmap (нижний ряд)
    ax3 = plt.subplot(gs[1, :])
    if feature_type == 'numerical':
        sns.boxplot(
            x=target, y=feature, data=df, ax=ax3,
            palette=['#4caf50', '#e74c3c'], showfliers=False
        )
        ax3.set_title(f"{feature} по классам {target}", fontsize=14)
        ax3.set_xlabel(target, fontsize=12)
        ax3.set_ylabel(feature, fontsize=12)
        ax3.set_xticklabels(['Лояльные', 'Ушедшие'])
        
        med = df.groupby(target)[feature].median()
        for i, m in enumerate(med):
            ax3.text(
                i, m, f"Медиана: {m:.2f}",
                ha='center', va='center', fontsize=12,
                bbox=dict(facecolor='gray', alpha=0.3)
            )
    else:
        heat = pd.crosstab(df[feature], df[target], normalize='index').fillna(0)
        sns.heatmap(
            heat, annot=True, fmt='.1%', cmap='coolwarm', ax=ax3,
            cbar_kws={'label': 'Доля'}
        )
        ax3.set_title(f"Тепловая карта: {feature} vs {target}", fontsize=14)
        ax3.set_xlabel("Отток", fontsize=12)
        ax3.set_ylabel(feature, fontsize=12)
        ax3.set_xticklabels(['Лояльные', 'Ушедшие'])
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

Определение какие признаки в какие разделы 

```python
def get_feature_types(df, threshold=10):
    """Определяет типы признаков с учетом пропусков"""
    cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    num_cols = df.select_dtypes(include=np.number).columns.tolist()
    
    # Для education будем использовать Label Encoding
    if 'education' in cat_cols:
        cat_cols.remove('education')
    
    # Разделение числовых признаков
    num_with_missing = [col for col in num_cols if df[col].isna().any()]
    num_without_missing = ['loan_amount', 'loan_term_months']  # как вы указали
    
    # Разделение категориальных признаков
    cat_with_missing = [col for col in cat_cols if df[col].isna().any()]
    cat_without_missing = [col for col in cat_cols if col not in cat_with_missing]
    
    return {
        'num_with_missing': num_with_missing,
        'num_without_missing': num_without_missing,
        'cat_ohe_with_missing': cat_with_missing,
        'cat_ohe_without_missing': cat_without_missing,
        'education_col': ['education'] if 'education' in df.columns else []
    }
```

Получаем разделы для pipeline 

```python
feature_types = get_feature_types(X)
num_with_missing = feature_types['num_with_missing']
num_without_missing = feature_types['num_without_missing']
cat_ohe_with_missing = feature_types['cat_ohe_with_missing']
cat_ohe_without_missing = feature_types['cat_ohe_without_missing']
education_col = feature_types['education_col']
feature_types
```

```python
# LabelEncoder
class CustomLabelEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.le = LabelEncoder()
        
    def fit(self, X, y=None):
        X_filled = np.where(pd.isnull(X), 'Unknown', X)
        self.le.fit(X_filled)
        return self
        
    def transform(self, X):
        X_filled = np.where(pd.isnull(X), 'Unknown', X)
        return self.le.transform(X_filled).reshape(-1, 1)

# Строим ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    # Числовые с пропусками
    ('num_missing', Pipeline([
        ('imputer', KNNImputer()),
        ('scaler', MinMaxScaler())
    ]), num_with_missing),
    
    # Числовые без пропусков
    ('num_no_missing', MinMaxScaler(), num_without_missing),
    
    # Категориальные для OneHot с пропусками
    ('cat_ohe_missing', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ]), cat_ohe_with_missing),
    
    # Категориальные для OneHot без пропусков
    ('cat_ohe_no_missing', OneHotEncoder(handle_unknown='ignore', sparse_output=False), 
     cat_ohe_without_missing),
    
    # Образование (LabelEncoder)
    ('education_encoder', CustomLabelEncoder(), education_col)
], remainder='drop')  # меняем на 'drop' чтобы избежать ошибки с remainder

# Pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

# Применяем преобразования
X_np = pipeline.fit_transform(X)
```

Исправленная функция восстановления имен столбцов

```python
def get_transformed_columns(transformer, original_columns):
    new_columns = []
    
    for name, trans, cols in transformer.transformers_:
        if name.startswith('cat_ohe'):
            encoder = trans.named_steps['onehot'] if isinstance(trans, Pipeline) else trans
            new_columns.extend(encoder.get_feature_names_out(cols))
        elif name in ['num_missing', 'num_no_missing', 'education_encoder']:
            new_columns.extend(cols)
        # Убрана обработка remainder, так как мы используем remainder='drop'
    
    return new_columns

# Получаем имена столбцов
output_cols = get_transformed_columns(pipeline.named_steps['preprocessor'], X.columns)

# Создаем DataFrame
X_ready = pd.DataFrame(X_np, columns=output_cols, index=X.index)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
import pandas as pd

def recursive_feature_elimination(
    X: pd.DataFrame,
    y: pd.Series,
    model_type: str = 'logistic',
    scoring: str = 'f1',
    cv: int = 5,
    random_state: int = 42,
    verbose: bool = True
) -> list:
    """
    Рекурсивно удаляет признаки, ухудшающие качество модели.
    
    Параметры:
    ----------
    X : pd.DataFrame
        Матрица признаков
    y : pd.Series
        Целевая переменная
    model_type : str, по умолчанию 'logistic'
        Тип модели: 'logistic' или 'random_forest'
    scoring : str, по умолчанию 'f1'
        Метрика для оценки
    cv : int, по умолчанию 5
        Количество фолдов для кросс-валидации
    random_state : int, по умолчанию 42
        Seed для воспроизводимости
    verbose : bool, по умолчанию True
        Выводить ли подробную информацию
        
    Возвращает:
    -----------
    list
        Список признаков, которые были удалены
    """
    
    # Инициализация модели
    if model_type == 'logistic':
        model = LogisticRegression(max_iter=1000, solver='lbfgs',
                                 random_state=random_state, n_jobs=-1)
        model_name = "Логистическая регрессия"
    elif model_type == 'random_forest':
        model = RandomForestClassifier(n_estimators=100,
                                     random_state=random_state, n_jobs=-1)
        model_name = "Случайный лес"
    else:
        raise ValueError("Неизвестный тип модели. Выберите 'logistic' или 'random_forest'")
    
    skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=random_state)
    remaining_features = list(X.columns)
    removed_features = []
    iteration = 1
    
    while True:
        if verbose:
            print(f"\n{'='*60}")
            print(f"Итерация {iteration}")
            print(f"Текущее количество признаков: {len(remaining_features)}")
        
        # Базовый score с текущими признаками
        base_score = cross_val_score(model, X[remaining_features], y,
                                   cv=skf, scoring=scoring, n_jobs=-1).mean()
        
        if verbose:
            print(f"Текущий {scoring} score: {base_score:.4f}")
        
        # Поиск признака для удаления
        worst_feature = None
        best_delta = 0
        
        for feature in remaining_features:
            temp_features = [f for f in remaining_features if f != feature]
            if not temp_features:  # Чтобы не удалить все признаки
                continue
                
            temp_score = cross_val_score(model, X[temp_features], y,
                                      cv=skf, scoring=scoring, n_jobs=-1).mean()
            delta = temp_score - base_score
            
            if delta > best_delta:
                best_delta = delta
                worst_feature = feature
        
        # Критерий остановки
        if worst_feature is None or best_delta <= 0:
            if verbose:
                print("\nНе найдено признаков для удаления, которые улучшают модель")
            break
            
        # Удаляем признак
        remaining_features.remove(worst_feature)
        removed_features.append(worst_feature)
        
        if verbose:
            print(f"\nУдален признак: {worst_feature}")
            print(f"Улучшение {scoring}: +{best_delta:.4f}")
            print(f"Новый {scoring} score: {base_score + best_delta:.4f}")
            print(f"Всего удалено признаков: {len(removed_features)}")
        
        iteration += 1
    
    if verbose:
        print("\n" + "="*60)
        print("Финальные результаты:")
        print(f"Изначальное количество признаков: {len(X.columns)}")
        print(f"Удалено признаков: {len(removed_features)}")
        print(f"Оставшееся количество признаков: {len(remaining_features)}")
        print("\nУдаленные признаки:")
        for i, feat in enumerate(removed_features, 1):
            print(f"{i}. {feat}")
        print("="*60)
    
    return removed_features

# Пример использования:
# Для логистической регрессии
removed_features_lr = recursive_feature_elimination(X_final, y, model_type='logistic')

# Для случайного леса
removed_features_rf = recursive_feature_elimination(X_final, y, model_type='random_forest')
```
